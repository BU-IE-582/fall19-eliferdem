## 3 - APPROACH

A ) DATA PREPROCESSING

Related data has been pulled from matches data (Finished matches of English Premiere League)
Duplicate values were eliminated.
Total scores of each match was counted.
Match results were flagged as: 
•	Home Team Wins
•	Tie
•	Away Team Win
Match statistic of Corner Kicks and Shots on Goal were added into data for each team and each game.
The bookmakers who has the least number of bets (least 5 bookmakers) were eliminated from the data.
Only related bet types were held in the data.

B ) FEATURE EXTRACTION

In order to predict upcoming matches, we used the information about teams’ previous performances. 
It is known from various researches that the home / away position of team affects players’ performance.
Thus, we extracted matches for each home team that they played as home team again Also, for each away team we extracted matches that they played as away team again.
Number of goals for each team from last 3 of these matches and last 5 of these matches were counted as features for model.

•	Mean of goal number of home team in last 3 matches that they are home team
•	Mean of goals number of home team in last 5 matches that they are home team
•	Mean of goals number of away team in last 3 matches that they are away team
•	Mean of goals number of away team in last 5 matches that they are away team

Shots on goals and corner kicks were used another meaningful indicator of teams’ general performance. Thus, we extracted those statistics to use in prediction.

•	Average number of shots on goal of home team in last 3 matches 
•	Average number of shots on goal of home team in last 5 matches 
•	Average number of shots on goal of away team in last 3 matches 
•	Average number of shots on goal of away team in last 5 matches 
•	Average number of corner kicks of home team in last 3 matches 
•	Average number of corner kicks of home team in last 5 matches 
•	Average number of corner kicks of away team in last 3 matches 
•	Average number of corner kicks of away team in last 5 matches

We used the total average number of scores for identical matches as another feature to predict the upcoming matches.
We used the odd data from bookmakers by calculating average of the normalized probabilities of match results (home win, tie and away win) from the given bets for each match. As we couldn’t find bet from each bookmaker - each match, in order to keep as much information as we can, we filled missing points with mean value of given bets.
FIFA team rating is an evaluation system that allows users to compare different teams by measuring how good, a team can play according to its players’ performance. We extracted FIFA Ratings of teams from the website (https://www.fifaindex.com) for each team. There are 4 ratings for each team. We used the difference between home and away team’s ratings as features.
We came up with 4 features as:

1.	Attack Difference as (Home Attack - Away Attack)
2.	Midfield Difference as (Home Midfield -Away Midfield)
3.	Overall Difference as (Home Overall -Away Overall)
4.	Defense Difference as (Home Defense -Away Defense)
The ELO rating system is a method for calculating the relative skill levels of players in zero-sum games, for instance, international football matches. We extracted these ratings for each team from the website (http://clubelo.com/ENG). 

## 4 - RESULTS
In our previous assignment, we worked on different prediction methods with the same data.
Stochastic Gradient Boosting Classifier brought better results than Penalized Regression Approach, Decision Tree and Random Forest. Thus, we fit our model using Stochastic Gradient Boosting Classifier and calculated the probabilities for each class (home win, tie, away win) as prediction result.
Stochastic Gradient Boosting Classifier gave better results in terms of RPS of test data. After tuning the parameters Stochastic Gradient Boosting Classifier model is improved. While trying different methods train error is also calculated to observe the relationship between train error and test error.
